```python

# Баганын нэрийг 'жолооч' гэж үзье. Та өөрийнхөөрөө солиорой.
# 1. Хоосон утгуудыг (NaN) түр хугацаанд хоосон текст '' болгож хөрвүүлэх (Алдаа заахаас сэргийлнэ)
zah_huu_df['Жолооч'] = zah_huu_df['Жолооч'].fillna('')

# 2. Бүх өгөгдлийг текст төрөл рүү хөрвүүлэх
zah_huu_df['Жолооч'] = zah_huu_df['Жолооч'].astype(str)

# 3. Текстүүдийн урд, ард байгаа илүүдэл хоосон зай болон "\n" (шинэ мөр)-ийг устгах
zah_huu_df['Жолооч'] = zah_huu_df['Жолооч'].str.strip()

# 4. Бүх нэрсийг ижилхэн "Эхний үсэг нь том, бусад нь жижиг" болгох (Capitalize)
# Жишээ нь: 'батболд', 'БАТБОЛД' хоёулаа -> 'Батболд' болно.
zah_huu_df['Жолооч'] = zah_huu_df['Жолооч'].str.capitalize()

# 5. Үг үсгийн зөрүүтэй эсвэл товчилсон нэрсийг ижилсүүлж солих (Mapping хийх)
# Та өөрийн датанд тааруулан доорх жагсаалтыг нэмж баяжуулж болно
солих_нэрс = {
    'Яабраа': 'Ябраа',
    'Үүүний': 'Үүний',
    'Мз': 'МЗ',
    'Э/болд': 'Энхболд',
    'Эболд': 'Энхболд',
    'Насмрай': 'Намсрай',
    'Батэрэдэнэ': 'Батэрдэнэ',
    'Батэрдэнэ\n': 'Батэрдэнэ',
    'Азаа2': 'Азбилэг',
    '99335584': 'Хүрлээ',
    'амаглан': 'Амгалан', 
    'араас мэндээ': 'Мэндсайхан', 
    'ээгий араас': 'Энхболд',
    'араас ээгий': 'Энхболд',
    'туул2': 'Ариунтуул',
    'Араас мэндээ': 'Мэндсайхан',
    'Араас батболд': 'Батболд',
    'Зоорий': 'Ариунзориг',
    'Араас ээгий': 'Ээгий', 
    'Туул2': 'Ариунтуул',
    'Ээгий араас': 'Ээгий',
    'Амаглан': 'Амгалан',
    'Мэндээ': 'Мэндсайхан',
    'Туул': 'Ариунтуул',
    'Зоригоо': 'Ариунзориг',
    'Пүрэвээ': 'Пүрэв',
    'Эндээ': 'Мэндсайхан',
    'Багана': 'Баганаа'

}
zah_huu_df['Жолооч'] = zah_huu_df['Жолооч'].replace(солих_нэрс)

# 6. Хэрэв жолооч баганад текст биш цэвэр утасны дугаар (99335584 гэх мэт) үлдсэн бол 
# түүнийг 'Тодорхойгүй' эсвэл хоосон болгох
zah_huu_df['Жолооч'] = zah_huu_df['Жолооч'].apply(lambda x: 'Тодорхойгүй' if x.isdigit() else x)

# Хамгийн сүүлд хоосон үлдсэн нүднүүдийг буцаагаад NaN болгож болно
zah_huu_df['Жолооч'] = zah_huu_df['Жолооч'].replace('', pd.NA)
```

```python


# 1. Жолооч багана нь хоосон (isna) байгаа мөрүүдийг олно.
# 2. Тэр мөрүүдийн 'утас' багананд шууд 0 утгыг оноож өгнө.
zah_huu_df.loc[zah_huu_df['Жолооч'].isna(), 'утас'] = 0

# (Нэмэлт) Хэрэв таны датанд жолооч багана дээр '<NA>' гэсэн текст өөрөө бичигдчихсэн байгаа бол:
zah_huu_df.loc[zah_huu_df['Жолооч'] == '<NA>', 'утас'] = 0
```

```python

# 1. Түр зуур 'утас' баганын урд ард байгаа хоосон зайг цэвэрлэх
# (Энэ нь " " гэсэн хоосон зайтай нүднүүдийг "" болгож хувиргана)
zah_huu_df['утас'] = zah_huu_df['утас'].astype(str).str.strip()

# 2. Утасны дугаар байхгүй мөрүүдийг шүүх:
# - Жинхэнэ хоосон (NaN) мөрүүд
# - Эсвэл текст болгоход 'nan' буюу хоосон үлдсэн мөрүүд
утасгүй_мөрүүд = zah_huu_df[zah_huu_df['утас'].isna() | (zah_huu_df['утас'] == '') | (zah_huu_df['утас'] == 'nan')]

# 3. Үр дүнг хэвлэж харах
print(f"=== Утасны дугаар байхгүй нийт {len(утасгүй_мөрүүд)} мөр олдлоо ===")
print(утасгүй_мөрүүд[['Жолооч', 'утас', 'барааны нэр']])
```

```python

# 1. 'барааны нэр' баганыг текст формат руу хөрвүүлэх
zah_huu_df['барааны нэр'] = zah_huu_df['барааны нэр'].fillna('').astype(str)

# 2. Нөхцөл: Барааны нэр дотор 'tengsu' гэсэн үг орсон байвал (case=False нь жижиг том үсэг хамаарахгүйг батална)
нөхцөл = zah_huu_df['барааны нэр'].str.contains('Эр бэлдмэл Tengsu-1,', case=False, na=False)

# 3. Тэр мөрүүдийн 'утас' багананд 99335584 дугаарыг тавих
zah_huu_df.loc[нөхцөл, 'утас'] = 99335584
```

```python
# Кодынхоо хамгийн доор үүнийг нэмээрэй:
print("=== 99335584 дугаараар шүүж шалгах ===")

# Утас багана нь тоо эсвэл текст байж болох тул хоёуланг нь шалгая
солигдсон_мөрүүд = zah_huu_df[(zah_huu_df['утас'] == 99335584) | (zah_huu_df['утас'] == '99335584')]

print(f"Нийт өөрчлөгдсөн мөрийн тоо: {len(солигдсон_мөрүүд)}")
print(солигдсон_мөрүүд[['барааны нэр', 'утас']])
```

```python

# 1. Түр хугацаанд шалгахын тулд текст төрөл рүү хөрвүүлнэ
zah_huu_df['опп'] = zah_huu_df['опп'].astype(str).str.strip()

# 2. 'опп' багана дотор огноо (2026) орчихсон байгаа мөрүүдийг (солигдсон мөрүүд) шүүх
солигдсон_нөхцөл = zah_huu_df['опп'].str.contains('2026', na=False)
солигдсон_дата = zah_huu_df[солигдсон_нөхцөл]

# 3. Хэдэн мөр солигдсоныг харах
print(f"Нийт багана нь солигдсон {len(солигдсон_дата)} мөр олдлоо.\n")

# 4. Тэдгээр солигдсон мөрүүдийг хэвлэж харах
print(солигдсон_дата[['опп', 'он сар']])
```

```python
# 1. Түр хугацаанд шалгахын тулд текст төрөл рүү хөрвүүлнэ
zah_huu_df['опп'] = zah_huu_df['опп'].astype(str).str.strip()

# 2. 'опп' багана дотор огноо (2026) орчихсон байгаа мөрүүдийг (солигдсон мөрүүд) шүүх
солигдсон_нөхцөл = zah_huu_df['опп'].str.contains('2026', na=False)
солигдсон_дата = zah_huu_df[солигдсон_нөхцөл]

# 3. Хэдэн мөр солигдсоныг харах
print(f"Нийт багана нь солигдсон {len(солигдсон_дата)} мөр олдлоо.\n")

# 4. Тэдгээр солигдсон мөрүүдийг хэвлэж харах
print(солигдсон_дата[['опп', 'он сар']])
```

```python

import numpy as np

# 1. Анхны огноог текст болгож, урд ард талын хоосон зайг цэвэрлэх
zah_huu_df['он сар'] = zah_huu_df['он сар'].astype(str).str.strip()

# 2. Жинхэнэ хоосон болон 'nan' гэж орсон текстүүдийг Pandas-ийн таньдаг ЖИНХЭНЭ ХООСОН (None/NaN) утга руу хөрвүүлэх
# Ингэж байж .ffill() функц хоосон нүдийг олж танина
zah_huu_df['он сар'] = zah_huu_df['он сар'].replace(['nan', 'None', ''], np.nan)

# 3. '1/3' гэх мэт алдаатайг Datetime-аас өмнө засах
zah_huu_df['он сар'] = zah_huu_df['он сар'].replace('1/3', '2026-01-03')

# 4. 🔥 СҮҮДЭРТЭХ АРГА (Forward Fill): Хоосон нүднүүдийг яг ДЭЭД талынх нь огноогоор нөхөх
zah_huu_df['он сар'] = zah_huu_df['он сар'].ffill()

# 5. Одоо бүх хоосон нүд нөхөгдсөн тул жинхэнэ Datetime руу хөрвүүлнэ
zah_huu_df['он сар'] = pd.to_datetime(zah_huu_df['он сар'], errors='coerce')

# 6. Цаг минутыг устгаж, зөвхөн ОГНОО (Жил-Сар-Өдөр) болгох
zah_huu_df['он сар'] = zah_huu_df['он сар'].dt.date

# 7. Үр дүнг шалгах
print("=== Хоосон огноонуудыг дээрх утгаар нь нөхсөн байдал ===")
print(zah_huu_df[['опп', 'он сар']].head(20))
```

```python
import pandas as pd
import streamlit as st
from datetime import datetime, timedelta

# Түрүүн цэвэрлэж бэлдсэн датафрейм тань 'zah_huu_df' нэртэй байгаа гэж үзье.
# Жишээ болгож огнооны баганыг бэлдэх (таны жинхэнэ огнооны баганы нэрээр солиорой):
if 'огноо' in zah_huu_df.columns:
    zah_huu_df['огноо_өдөр'] = pd.to_datetime(zah_huu_df['огноо']).dt.date
else:
    # Хэрэв огнооны багана байхгүй бол өнөөдрөөр түр дүүргэв (тест хийхэд зориулж)
    zah_huu_df['огноо_өдөр'] = datetime.today().date()

st.title("💬 Захиалгын Аналитик Чатбот")
st.write("Та захиалгын төлөв, тоог өдрөөр нь асуух боломжтой. (Өнөөдөр, Өчигдөр гэх мэт)")

# Чатны түүхийг хадгалах Сессийн төлөв (Session State)
if "messages" not in st.session_state:
    st.session_state.messages = [
        {"role": "assistant", "content": "Сайн байна уу! Би захиалга мэдээллийн туслах байна. Та асуултаа үлдээнэ үү."}
    ]

# Өмнөх чатны түүхийг дэлгэцэнд хэвлэх
for message in st.session_state.messages:
    with st.chat_message(message["role"]):
        st.write(message["content"])

# Хэрэгэгчийн асуулт оруулах хэсэг
if user_query := st.chat_input("Асуултаа энд бичнэ үү... (ж.нь: Өнөөдөр хэдэн захиалга цуцлагдсан бэ?)"):
    
    # 1. Хэрэглэгчийн асуултыг дэлгэцэнд харуулах ба түүхэнд хадгалах
    with st.chat_message("user"):
        st.write(user_query)
    st.session_state.messages.append({"role": "user", "content": user_query})
    
    # --- АНАЛИЗ ХИЙХ, ХАРИУЛТ БЭЛДЭХ ЛОГИК ---
    query_lower = user_query.lower()
    
    # A. Огноог тогтоох
    target_date = datetime.today().date() # Дефолт нь өнөөдөр
    date_text = "Өнөөдөр"
    
    if "өчигдөр" in query_lower:
        target_date = datetime.today().date() - timedelta(days=1)
        date_text = "Өчигдөр"
    elif "уржигдар" in query_lower:
        target_date = datetime.today().date() - timedelta(days=2)
        date_text = "Уржигдар"
        
    # Б. Төлөвийг тогтоох (Таны ангилсан 4 төлөв)
    status_target = None
    if "Хүргэсэн" in query_lower or "амжилттай" in query_lower:
        status_target = "Хүргэсэн"
    elif "Цуцалсан" in query_lower or "цуцлагдсан" in query_lower:
        status_target = "Цуцалсан"
    elif 'Хойшлогдсон захиалга' in query_lower or "хойшилсон" in query_lower:
        status_target = "Хойшлогдсон захиалга"
    elif "Буцаасан" in query_lower or "буцаалт" in query_lower:
        status_target = "Буцаасан"

    # В. Датафреймээс шүүж хариулт үүсгэх
    # Тухайн өдрийн датаг шүүх
    daily_df = zah_huu_df[zah_huu_df['огноо_өдөр'] == target_date]
    total_daily = len(daily_df)
    
    if status_target:
        # Хэрэв тодорхой нэг төлөв асуусан бол (ж.нь: өнөөдөр хэдэн хүргэсэн бэ?)
        count = len(daily_df[daily_df['мэдээ'] == status_target])
        bot_response = f"📊 **{date_text}** ({target_date}) нийт **{count}** захиалга **'{status_target}'** төлөвтэй байна."
    else:
        # Хэрэв ерөнхийд нь асуусан бол (ж.нь: өнөөдөр ямар байна, өнөөдрийн захиалга хэд байна?)
        summary = daily_df['мэдээ'].value_counts()
        
        bot_response = f"📊 **{date_text}** ({target_date}) нийт **{total_daily}** захиалга орсон байна.\n\n"
        bot_response += f"• **Хүргэсэн:** {summary.get('Хүргэсэн', 0)}ш\n"
        bot_response += f"• **Хойшлогдсон:** {summary.get('Хойшлогдсон захиалга', 0)}ш\n"
        bot_response += f"• **Буцаасан:** {summary.get('Буцаасан', 0)}ш\n"
        bot_response += f"• **Цуцалсан:** {summary.get('Цуцалсан', 0)}ш"

    # 2. Бот-ын хариултыг дэлгэцэнд харуулах ба түүхэнд хадгалах
    with st.chat_message("assistant"):
        st.write(bot_response)
    st.session_state.messages.append({"role": "assistant", "content": bot_response})
```

```python
import pandas as pd
import streamlit as st
import re
from datetime import datetime, timedelta

# 1. Огноо таних функц (Хэрэглэгчийн бичсэн текстээс огноог шүүж авна)
def extract_date_from_text(text):
    # '2026-01-03' эсвэл '2026-1-3' эсвэл '2026/01/03' гэсэн форматуудыг хайх regex
    date_pattern = r'(\d{4})[-/.](\d{1,2})[-/.](\d{1,2})'
    match = re.search(date_pattern, text)
    
    if match:
        year, month, day = match.groups()
        # Огноог стандарт datetime.date объект болгож хувиргах
        return datetime(int(year), int(month), int(day)).date()
    return None

# --- Стримлит Чатбот Интерфэйс ---
st.title("💬 Online Chatbot")
st.write("Захиалгын мэдээллийг тодорхой огноогоор эсвэл ерөнхий өдрөөр асуугаарай. (ж.нь: 2026-01-03-ний өдөр нийт захиалга хэд байна?)")

# Чатны түүхийг хадгалах Сессийн төлөв
if "messages" not in st.session_state:
    st.session_state.messages = [
        {"role": "assistant", "content": "Сайн байна уу! Би захиалгын аналитик туслах байна. Та асуултаа үлдээнэ үү."}
    ]

# Өмнөх чатны түүхийг дэлгэцэнд хэвлэх
for message in st.session_state.messages:
    with st.chat_message(message["role"]):
        st.write(message["content"])

# Хэрэглэгчээс асуулт авах
if user_query := st.chat_input("Асуултаа бичнэ үү..."):
    
    # Хэрэглэгчийн асуултыг харуулах ба хадгалах
    with st.chat_message("user"):
        st.write(user_query)
    st.session_state.messages.append({"role": "user", "content": user_query})
    
    # --- АНАЛИЗ ХИЙХ БУЮУ ӨДӨР ТАНИХ ЛОГИК ---
    query_lower = user_query.lower()
    
    # Асуултаас шууд огноо бичсэн эсэхийг шалгах (ж.нь: 2026-01-03)
    target_date = extract_date_from_text(user_query)
    
    if target_date:
        date_text = str(target_date)
    else:
        # Хэрэв тодорхой огноо бичээгүй бол "өнөөдөр", "өчигдөр"-ийг шалгах
        if "өчигдөр" in query_lower:
            target_date = datetime.today().date() - timedelta(days=1)
            date_text = "Өчигдөр"
        elif "уржигдар" in query_lower:
            target_date = datetime.today().date() - timedelta(days=2)
            date_text = "Уржигдар"
        else:
            target_date = datetime.today().date()
            date_text = "Өнөөдөр"

    # --- ДАТА ШҮҮЖ ХАРИУЛТ БЭЛДЭХ ---
    # Таны датафрейм дээрх огнооны баганыг 'огноо_өдөр' гээд зөвхөн Date форматтай болгосон байна гэж үзлээ
    if 'огноо_өдөр' not in zah_huu_df.columns:
        zah_huu_df['огноо_өдөр'] = pd.to_datetime(zah_huu_df['мэдээ'], errors='coerce').dt.date 
        # Тэмдэглэл: Та огноог 'мэдээ' баганаас шууд авч байгаа бол дээрх шиг хөрвүүлнэ, 
        # эсвэл тусдаа огнооны багана байгаа бол түүнийгээ хөрвүүлээрэй.

    # Тухайн өдрийн датаг шүүх
    daily_df = zah_huu_df[zah_huu_df['огноо_өдөр'] == target_date]
    total_orders = len(daily_df)
    
    # Төлөв тус бүрээр тоолох
    summary = daily_df['мэдээ'].value_counts()
    
    # Хариултын текстийг таны хүссэн яг таг форматаар бэлдэх
    bot_response = f"📅 **{date_text}**-ний өдрийн захиалгын нэгдсэн мэдээлэл:\n\n"
    bot_response += f"• **Нийт захиалга:** {total_orders}ш\n"
    bot_response += f"--- 分割 ---\n" # Зааглагч шугам
    bot_response += f"• 🟩 **Хүргэсэн:** {summary.get('Хүргэсэн', 0)}ш\n"
    bot_response += f"• 🟨 **Хойшлогдсон захиалга:** {summary.get('Хойшлогдсон захиалга', 0)}ш\n"
    bot_response += f"• 🟧 **Буцаасан:** {summary.get('Буцаасан', 0)}ш\n"
    bot_response += f"• 🟥 **Цуцалсан:** {summary.get('Цуцалсан', 0)}ш"

    # Ботын хариултыг харуулах ба хадгалах
    with st.chat_message("assistant"):
        st.write(bot_response)
    st.session_state.messages.append({"role": "assistant", "content": bot_response})
```

```python
import pandas as pd
import streamlit as st
import re
from datetime import datetime, timedelta

# 1. Огноо таних функц (Хэрэглэгчийн бичсэн текстээс огноог шүүж авна)
def extract_date_from_text(text):
    # '2026-01-03' эсвэл '2026-1-3' эсвэл '2026/01/03' гэсэн форматуудыг хайх regex
    date_pattern = r'(\d{4})[-/.](\d{1,2})[-/.](\d{1,2})'
    match = re.search(date_pattern, text)
    
    if match:
        year, month, day = match.groups()
        # Огноог стандарт datetime.date объект болгож хувиргах
        return datetime(int(year), int(month), int(day)).date()
    return None

# --- Стримлит Чатбот Интерфэйс ---
st.title("💬 CF - Тех Карт & Аналитик Chatbot")
st.write("Захиалгын мэдээллийг тодорхой огноогоор эсвэл ерөнхий өдрөөр асуугаарай. (ж.нь: 2026-01-03-ний өдөр нийт захиалга хэд байна?)")

# Чатны түүхийг хадгалах Сессийн төлөв
if "messages" not in st.session_state:
    st.session_state.messages = [
        {"role": "assistant", "content": "Сайн байна уу! Би захиалгын аналитик туслах байна. Та асуултаа үлдээнэ үү."}
    ]

# Өмнөх чатны түүхийг дэлгэцэнд хэвлэх
for message in st.session_state.messages:
    with st.chat_message(message["role"]):
        st.write(message["content"])

# Хэрэглэгчээс асуулт авах
if user_query := st.chat_input("Асуултаа бичнэ үү..."):
    
    # Хэрэглэгчийн асуултыг харуулах ба хадгалах
    with st.chat_message("user"):
        st.write(user_query)
    st.session_state.messages.append({"role": "user", "content": user_query})
    
    # --- АНАЛИЗ ХИЙХ БУЮУ ӨДӨР ТАНИХ ЛОГИК ---
    query_lower = user_query.lower()
    
    # Асуултаас шууд огноо бичсэн эсэхийг шалгах (ж.нь: 2026-01-03)
    target_date = extract_date_from_text(user_query)
    
    if target_date:
        date_text = str(target_date)
    else:
        # Хэрэв тодорхой огноо бичээгүй бол "өнөөдөр", "өчигдөр"-ийг шалгах
        if "өчигдөр" in query_lower:
            target_date = datetime.today().date() - timedelta(days=1)
            date_text = "Өчигдөр"
        elif "уржигдар" in query_lower:
            target_date = datetime.today().date() - timedelta(days=2)
            date_text = "Уржигдар"
        else:
            target_date = datetime.today().date()
            date_text = "Өнөөдөр"

    # --- ДАТА ШҮҮЖ ХАРИУЛТ БЭЛДЭХ ---
    # Таны датафрейм дээрх огнооны баганыг 'огноо_өдөр' гээд зөвхөн Date форматтай болгосон байна гэж үзлээ
    if 'огноо_өдөр' not in zah_huu_df.columns:
        zah_huu_df['огноо_өдөр'] = pd.to_datetime(zah_huu_df['мэдээ'], errors='coerce').dt.date 
        # Тэмдэглэл: Та огноог 'мэдээ' баганаас шууд авч байгаа бол дээрх шиг хөрвүүлнэ, 
        # эсвэл тусдаа огнооны багана байгаа бол түүнийгээ хөрвүүлээрэй.

    # Тухайн өдрийн датаг шүүх
    daily_df = zah_huu_df[zah_huu_df['огноо_өдөр'] == target_date]
    total_orders = len(daily_df)
    
    # Төлөв тус бүрээр тоолох
    summary = daily_df['мэдээ'].value_counts()
    
    # Хариултын текстийг таны хүссэн яг таг форматаар бэлдэх
    bot_response = f"📅 **{date_text}**-ний өдрийн захиалгын нэгдсэн мэдээлэл:\n\n"
    bot_response += f"• **Нийт захиалга:** {total_orders}ш\n"
    bot_response += f"--- 分割 ---\n" # Зааглагч шугам
    bot_response += f"• 🟩 **Хүргэсэн:** {summary.get('Хүргэсэн', 0)}ш\n"
    bot_response += f"• 🟨 **Хойшлогдсон захиалга:** {summary.get('Хойшлогдсон захиалга', 0)}ш\n"
    bot_response += f"• 🟧 **Буцаасан:** {summary.get('Буцаасан', 0)}ш\n"
    bot_response += f"• 🟥 **Цуцалсан:** {summary.get('Цуцалсан', 0)}ш"

    # Ботын хариултыг харуулах ба хадгалах
    with st.chat_message("assistant"):
        st.write(bot_response)
    st.session_state.messages.append({"role": "assistant", "content": bot_response})
```
